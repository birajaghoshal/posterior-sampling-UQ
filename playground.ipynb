```python
from discriminator.dis import Dis
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_gt = torch.load('./data/test_gt.pt').float()
test_observed = torch.load('./data/test_observed.pt').float()
training_gt = torch.load('./data/training_gt.pt').float()
training_observed = torch.load('./data/training_observed.pt').float()
```

```python
test_gt = torch.sqrt(test_gt[:,0]**2+test_gt[:,1]**2).reshape((300,1,256,256))
test_observed = torch.sqrt(test_observed[:,0]**2+test_observed[:,1]**2).reshape((300,1,256,256))
training_gt = torch.sqrt(training_gt[:,0]**2+training_gt[:,1]**2).reshape((600,1,256,256))
training_observed = torch.sqrt(training_observed[:,0]**2+training_observed[:,1]**2).reshape((600,1,256,256))
```

```python
torch.save(test_gt, './new_data/test.gt.pt')
torch.save(test_observed, './new_data/test_observed.pt')
torch.save(training_gt, './new_data/training_gt.pt')
torch.save(training_observed, './new_data/training_observed.pt')
```
